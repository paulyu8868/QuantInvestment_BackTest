```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

import os
import matplotlib.pyplot as plt
from matplotlib import font_manager
import matplotlib.font_manager as fm

fe = fm.FontEntry(
    fname=r'/usr/share/fonts/truetype/nanum/NanumGothic.ttf', # ttf    
    name='NanumGothic')                        #   
fm.fontManager.ttflist.insert(0, fe)              # Matplotlib  

plt.rcParams.update({'font.size': 15, 'font.family': 'NanumGothic'}) #  


# Create DataFrame
data = {
    '': ['', ' (6)', ' (7)', ' (8)', '', '',
              '', '(8)', '(9)', '(10)'],
    '': [68.3, 86.4, 164.3, 157.4, 161.8, 129.9, 144.8, 175.6, 154.1, 134.7],
    'MDD': [-29.4, -54.1, -39.6, -34.5, -40.8, -53.0, -28.1, -38.7, -34.0, -30.6]
}

df = pd.DataFrame(data)

# Create scatter plot
plt.figure(figsize=(15, 8))

plt.scatter(df[''], df['MDD'], alpha=0.6, s=100)

# Add labels for each point
for i, txt in enumerate(df['']):
    plt.annotate(txt, (df[''][i], df['MDD'][i]), xytext=(5,5), textcoords='offset points')

# Add diagonal line
min_val = min(min(df['']), min(df['MDD']))
max_val = max(max(df['']), max(df['MDD']))
plt.plot([min_val, max_val], [min_val, max_val], 'k--', alpha=0.3)

# Customize plot
plt.xlabel(' (%)')
plt.ylabel('MDD (%)')
plt.title('TQQQ  (2020-2023)')
plt.grid(True, alpha=0.3)

# Set axis limits
plt.xlim(60, 180)
plt.ylim(-60, -25)

plt.tight_layout()
plt.show()
```
