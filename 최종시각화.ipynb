```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

import os
import matplotlib.pyplot as plt
from matplotlib import font_manager
import matplotlib.font_manager as fm

fe = fm.FontEntry(
    fname=r'/usr/share/fonts/truetype/nanum/NanumGothic.ttf', # ttf 파일이 저장되어 있는 경로
    name='NanumGothic')                        # 원하는 폰트 설정
fm.fontManager.ttflist.insert(0, fe)              # Matplotlib에 폰트 추가

plt.rcParams.update({'font.size': 15, 'font.family': 'NanumGothic'}) # 폰트 설정


# Create DataFrame
data = {
    '매매법': ['무한매수법', '떨사오팔 (6분할)', '떨사오팔 (7분할)', '떨사오팔 (8분할)', '종사종팔', '동파법',
              '강화매매법', '침몰방지법(8분할)', '침몰방지법(9분할)', '침몰방지법(10분할)'],
    '수익률': [68.3, 86.4, 164.3, 157.4, 161.8, 129.9, 144.8, 175.6, 154.1, 134.7],
    'MDD': [-29.4, -54.1, -39.6, -34.5, -40.8, -53.0, -28.1, -38.7, -34.0, -30.6]
}

df = pd.DataFrame(data)

# Create scatter plot
plt.figure(figsize=(15, 8))

plt.scatter(df['수익률'], df['MDD'], alpha=0.6, s=100)

# Add labels for each point
for i, txt in enumerate(df['매매법']):
    plt.annotate(txt, (df['수익률'][i], df['MDD'][i]), xytext=(5,5), textcoords='offset points')

# Add diagonal line
min_val = min(min(df['수익률']), min(df['MDD']))
max_val = max(max(df['수익률']), max(df['MDD']))
plt.plot([min_val, max_val], [min_val, max_val], 'k--', alpha=0.3)

# Customize plot
plt.xlabel('수익률 (%)')
plt.ylabel('MDD (%)')
plt.title('TQQQ 비교 (2020-2023)')
plt.grid(True, alpha=0.3)

# Set axis limits
plt.xlim(60, 180)
plt.ylim(-60, -25)

plt.tight_layout()
plt.show()
```
