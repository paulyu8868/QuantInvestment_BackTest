```python

plt.rcParams.update({'font.size': 23, 'font.family': 'NanumGothic'}) #폰트설정

# # Create DataFrame
# data = {
#     '매매법': ['무한매수법', '떨사오팔 (8분할)', '종사종팔', '동파법',
#               '강화매매법', '침몰방지법(8분할)'],
#     '수익률': [68.3, 157.4, 161.8, 129.9, 144.8, 175.6],
#     'MDD': [-29.4, -34.5, -40.8, -53.0, -28.1, -38.7]
# }

# Create DataFrame
data = {
    '매매법': ['매매법A', '매매법B', '매매법C', '매매법D',
              '강화매매법', '침몰방지법'],
    '수익률': [68.3, 157.4, 161.8, 129.9, 144.8, 175.6],
    'MDD': [-29.4, -34.5, -40.8, -53.0, -28.1, -38.7]
}

df = pd.DataFrame(data)

# 색상 설정 (침몰방지법과 강화매매법만 파란색으로 설정)
colors = ['blue' if method in ['침몰방지법', '강화매매법'] else 'gray'
          for method in df['매매법']]

# 라벨 색상 설정
label_colors = ['blue' if method in ['침몰방지법', '강화매매법'] else 'black'
                for method in df['매매법']]

# 산점도 그리기
plt.figure(figsize=(15, 8))
plt.scatter(df['수익률'], df['MDD'], alpha=0.6, s=100, c=colors)

# 데이터 라벨 추가 (글자 색상 변경)
for i, txt in enumerate(df['매매법']):
    plt.annotate(txt, (df['수익률'][i], df['MDD'][i]),
                 xytext=(5, 5), textcoords='offset points',
                 color=label_colors[i])  # 글자 색상 지정

# Add diagonal line
min_val = min(min(df['수익률']), min(df['MDD']))
max_val = max(max(df['수익률']), max(df['MDD']))
plt.plot([min_val, max_val], [min_val, max_val], 'k--', alpha=0.3)

# Customize plot
plt.xlabel('수익률 (%)')
plt.ylabel('MDD (%)')
plt.title('TQQQ (2020-2023)')
plt.grid(True, alpha=0.3)

# Set axis limits
plt.xlim(60, 200)
plt.ylim(-60, -25)

plt.tight_layout()
plt.show()
```
